#Task 7: Merging and Joining DataFrame

```python
#merge two DataFrames on common column
df = pd.read_csv('messed_dataset.csv')
df
```

```python
import pandas as pd
df1 = pd.DataFrame({'Name':['Monu'],
                    'age':[20],
                    'email':["monubabu@gmail.com"]})
df2 = pd.DataFrame({'Name':['Sonu'],
                    'age':[30],
                    'email':["sonubabu@gmail.com"]})
pd.merge(df1,df2,on='Name')
```

```python
#join two dataFrames using their indices and add suffixes to avoid column name conflicts
df1.join(df2, lsuffix='_left', rsuffix='_right')
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot')
plt.show()
```

```python
#scatter plot
plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
#bar plot
plt.bar(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar Plot')
plt.show()
```

```python
#Histogram
plt.hist(x)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Histogram')
plt.show()
```

```python
#boxplot
plt.boxplot(x)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Box Plot')
plt.show()
```
