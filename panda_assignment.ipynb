```python
import numpy as np
# replacing 'invalid' to 'NaN'
df['Salary'].replace('invalid',0,inplace=True)
df['Salary']=pd.to_numeric(df['Salary'], errors='coerce')
df['Salary'].dtypes
```

### Data Visualization

```python
# Groupby Operation
# Calculating the mean salary based on country
salary_based_on_country=df.groupby('Country')['Salary'].mean().reset_index()
print("Average Salary Based on Country:\n",salary_based_on_country)
```

```python
# Converting the Joining_Date to datetime

# Some of the row value of Joining_Date is invalid. So replace it 
df['Joining_Date'].replace('invalid','2022-01-01',inplace=True)
df['Joining_Date']=pd.to_datetime(df['Joining_Date'],errors='coerce')
df['Joining_Date']


# Extracting year,months and day from data
df['Year']=df['Joining_Date'].dt.year
df['Months']=df['Joining_Date'].dt.month
df['Day']=df['Joining_Date'].dt.day

df
```

```python
# Calculating the days since joining of the employee
df['Days_Since_Joining']=(pd.Timestamp('2024-05-07')- df['Joining_Date']).dt.days
df
```

### Merging and Joining Data

```python
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Name1', 'Name2', 'Name3', 'Name4']
})

df2 = pd.DataFrame({
    'ID': [1, 2, 3, 5],
    'Age': [25, 22, 24, 20]
})

merged_data_inner=pd.merge(df1,df2,on='ID',how='inner')
print('Inner Merged:\n',merged_data_inner)
merged_data_left=pd.merge(df1,df2,on='ID',how='left')
print('\n\n\n Left Merged:\n',merged_data_left)
merged_data_right=pd.merge(df1,df2,on='ID',how='right')
print('\n\n\n Right Merged:\n',merged_data_right)
```

- Joinig

```python
df3 = pd.DataFrame({
    'Name': ['Sujan', 'Sanjeev', 'Sunil', 'Sangita'],
    'Age': [29, 22, 20, 25]
}, index=[1, 2, 3, 4])

df4 = pd.DataFrame({
    'Salary': [50000, 60000, 70000, 80000]
}, index=[1, 2, 3, 5])
joined_df = df3.join(df4)
print('Joined Data:\n',joined_df)
```
