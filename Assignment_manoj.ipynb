# Task 7: Merging and Joining DataFrames

```python
data1 = {'Name': ['Ram', 'Hari', 'Sita', 'Gita'],
        'Age': [20, 21, 19, 18]}
```

```python
data2 = {'Call': ['Shyam', 'Laxman', 'Rita', 'Sarita'],
        'Roll': [27, 28, 15, 13]}
```

```python
jdf=pd.DataFrame(data1)
```

```python
jdf2=pd.DataFrame(data2)
```

```python
joined_df=jdf.join(jdf2)
joined_df
```

```python
pd.merge(joined_df, jdf2, left_index=True, right_index=True)
```

```python
pd.concat([jdf2, jdf], axis=1)
```
